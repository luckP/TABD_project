```python
df = pd.read_csv('offsets3.csv', header=None)
df.head()
```

```python
df = df.transform(lambda x: x.transform(lambda y: (float(y.split(' ')[0]), float(y.split(' ')[1]))) )
```

```python
arr = df.to_numpy()
```

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML

# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(0, 2), ylim=(-2, 2))
line, = ax.plot([], [], lw=2, marker='o', linestyle='')

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line,

# animation function.  This is called sequentially
def animate(i):
    x = [a[0] for a in arr[i] ]
    y = [a[1] for a in arr[i] ]
    line.set_data(x, y)
    if i%100 == 0:
        print('i: ' + str(i) + ' - ' str(len(arr)))
    return line,

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, frames=len(arr),
                               init_func=init,  repeat=False, interval=5)

# anim = animation.FuncAnimation(fig, animate, init_func=init,
#                                frames=200, interval=20, blit=True)

# save the animation as an mp4.  This requires ffmpeg or mencoder to be
# installed.  The extra_args ensure that the x264 codec is used, so that
# the video can be embedded in html5.  You may need to adjust this for
# your system: for more information, see
# http://matplotlib.sourceforge.net/api/animation_api.html
# plt.show()
plt.rcParams['animation.ffmpeg_path'] = '/usr/local/bin/ffmpeg'
HTML(anim.to_html5_video())
# anim.save('bloch_sphere.mp4', fps=20)
```
